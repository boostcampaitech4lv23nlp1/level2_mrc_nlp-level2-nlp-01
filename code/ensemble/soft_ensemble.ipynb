### hard voting

```python
target_dirs = ['./61.25', './62.92']
```

```python
import os
import itertools
from collections import defaultdict, Counter

mrc_dict = defaultdict(list)
for directory in target_dirs:
    df = pd.read_json(os.path.join(directory, 'nbest_predictions.json'))
    for id, top_k in df.items():
        candidate = itertools.chain(*[[value['text']] * (int(value['probability']*20)+1) for value in list(top_k.values)])
        mrc_dict[id].extend(candidate)

pred_dict = {}
for id, candidate in mrc_dict.items():
    pred = Counter(candidate).most_common(1)[0][0]
    pred_dict[id] = pred
output = pd.Series(pred_dict)

print(output)
with open('./ensemble.json', 'w+', encoding='utf-8') as f:
    output.to_json(f, force_ascii=False)
```
